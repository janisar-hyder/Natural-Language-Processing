```python
import cv2
import matplotlib.pyplot as plt

# Helper function to display original and processed images side by side
def display_side_by_side(original, processed, title):
    plt.figure(figsize=(12, 6))
    
    # Display original image
    plt.subplot(1, 2, 1)
    plt.title(f"Original - {title}")
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    # Display sharpened image
    plt.subplot(1, 2, 2)
    plt.title(f"Sharpened - {title}")
    plt.imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    
    # Show the plot
    plt.show()

# Load the image from a specific path
image_path = 'b12.jpg'
image = cv2.imread(image_path)

# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(image, (9, 9), 10)

# Sharpen the image by subtracting the blurred version from the original
sharpened_image = cv2.addWeighted(image, 1.5, blurred_image, -0.5, 0)

# Display the original and sharpened images side by side
display_side_by_side(image, sharpened_image, 'input.jpg')
```
